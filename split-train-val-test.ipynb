```python
import numpy as np
import pandas as pd
import os
```

```python
data = pd.read_csv('../deepnano.csv',index_col=0)
train = data.sample(frac=0.8)
print(train)
test = data.drop(train.index)
val = train.sample(frac=0.2)
train = train.drop(val.index)

train['status'] = 'train'
test['status'] = 'test'
val['status'] = 'val'

data = pd.concat([train,val,test]).sample(frac=1)
```

```python
data.to_csv('clean_data2.csv')
```
